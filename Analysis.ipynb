```python
# Load and Explore the Dataset
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# Load dataset from sklearn
iris = load_iris(as_frame=True)
df = iris.frame  # DataFrame version

# Add target (species names) to the DataFrame
df['species'] = df['target'].apply(lambda x: iris.target_names[x])

# Display first few rows
print("First 5 rows of dataset:")
df.head()
```

```python
# Check structure of dataset
print("\nDataset Info:")
print(df.info())

print("\nMissing values per column:")
print(df.isnull().sum())

# No missing values here, but weâ€™ll show how to handle if they exist:
# df = df.dropna()   # OR df.fillna(method='ffill')
```

```python
 #Basic Data Analysis
print("\nBasic Statistics:")
display(df.describe())

# Group by species and compute mean of numerical columns
species_means = df.groupby("species").mean()
print("\nAverage values per species:")
display(species_means)
```

```python
# Task 3: Data Visualization

# Line chart: Trend of sepal length values (just as an example, sorted)
plt.figure(figsize=(8,5))
plt.plot(sorted(df["sepal length (cm)"].values))
plt.title("Line Chart: Sorted Sepal Length Values")
plt.xlabel("Index")
plt.ylabel("Sepal Length (cm)")
plt.show()

# Bar chart: Average petal length per species
plt.figure(figsize=(8,5))
sns.barplot(x=species_means.index, y=species_means["petal length (cm)"], palette="viridis")
plt.title("Average Petal Length by Species")
plt.xlabel("Species")
plt.ylabel("Petal Length (cm)")
plt.show()

# Histogram: Distribution of sepal length
plt.figure(figsize=(8,5))
plt.hist(df["sepal length (cm)"], bins=15, color="skyblue", edgecolor="black")
plt.title("Histogram of Sepal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Frequency")
plt.show()

# Scatter plot: Sepal length vs. Petal length
plt.figure(figsize=(8,5))
sns.scatterplot(x="sepal length (cm)", y="petal length (cm)", hue="species", data=df, palette="Set1")
plt.title("Scatter Plot: Sepal Length vs Petal Length")
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Petal Length (cm)")
plt.legend(title="Species")
plt.show()

# Error Handling Example: Reading a CSV (uncomment for custom dataset)
"""
try:
    df_custom = pd.read_csv("your_dataset.csv")
    print("Dataset loaded successfully!")
except FileNotFoundError:
    print("Error: The file was not found. Please check the path.")
except pd.errors.ParserError:
    print("Error: Problem parsing the file. Please check the format.")
"""

# ------------------------------------------
# Findings:
# - Setosa species generally have smaller petal dimensions.
# - Virginica species tends to have the largest flower parts.
# - Sepal length distribution shows most flowers between 5 and 7 cm.
# - Strong positive relationship between sepal length and petal length.




```
